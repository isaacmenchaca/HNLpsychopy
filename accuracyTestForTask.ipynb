#### Extract actual trial-to-trial data

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dfLength = len(data['Experiment Data'])
data['Experiment Data'] = data['Experiment Data'].str.replace("\'", "\"")
data['Experiment Data'] = data['Experiment Data'].str.replace("[", "")
data['Experiment Data'] = data['Experiment Data'].str.replace("]", "")
data['Experiment Data'] = data['Experiment Data'].str.replace("None", "\"None\"")
data['Experiment Data'] = data['Experiment Data'].str.replace("True", "\"True\"")
data['Experiment Data'] = data['Experiment Data'].str.replace("False", "\"False\"")
```

```python
extractedData = [json.loads(trialInfo) for trialInfo in data['Experiment Data'].values ]
```

```python
extractedData
```

```python
numCorrect = 0
accuracy = []
probability = []

for i, dictTrialInfo in enumerate(extractedData):
    if dictTrialInfo['Stim Type'] == 'response':
        if dictTrialInfo['Correct'] == "True":
            numCorrect += 1
        accuracy.append(dictTrialInfo['Correct'])
        probability.append(round(dictTrialInfo['Probability of 0'], 5))
print("Percent Correct:", numCorrect/100)
```

```python
accuracyDF = pd.DataFrame({'probability of 0': probability, 'accuracy': accuracy})
accuracyDF['probability of 0'].unique()
```
