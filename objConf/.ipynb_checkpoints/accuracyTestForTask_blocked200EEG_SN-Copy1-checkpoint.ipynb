#### Extract actual trial-to-trial data

```python
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', None)
```

```python
dfLength = len(data['Experiment Data'])
data['Experiment Data'] = data['Experiment Data'].str.replace("\'", "\"")
data['Experiment Data'] = data['Experiment Data'].str.replace("[", "")
data['Experiment Data'] = data['Experiment Data'].str.replace("]", "")
data['Experiment Data'] = data['Experiment Data'].str.replace("None", "\"None\"")
data['Experiment Data'] = data['Experiment Data'].str.replace("True", "\"True\"")
data['Experiment Data'] = data['Experiment Data'].str.replace("False", "\"False\"")
extractedData = [json.loads(trialInfo) for trialInfo in data['Experiment Data'].values ]
```

```python
trialsDf = pd.DataFrame(extractedData)

trialsDf['Correct'] = trialsDf['Correct'].map({'True': True, 'False': False})
display(trialsDf)
```

#### Obtain insight by extracting response information

```python
responsesDf = trialsDf[trialsDf['Stim Type'] == 'response']
responsesDf = responsesDf.loc[:, ~responsesDf.columns.isin(['Instructions Start Time (ms)', 'CEDRUS Total Time (ms)'])]
print("Percent Correct:", responsesDf['Correct'].sum() / len(responsesDf['Correct']))
display(responsesDf)
```

start time here is not from the beginning of first sample but rather the timing of when cross hair appeared and when they decided to answer. can get time from the first sample though as that was recorded.

```python
total0sList = []
totalRTOpt = []
blocks = 6
trialsPerBlock = [50, 50, 50, 50, 50, 47]
for j in range(blocks):
    for i in trialsPerBlock:
        x0forResponse = trialsDf[(trialsDf['Block'] == j) & (trialsDf['Trial'] == i) & (trialsDf['Stim Type'] == 'X0')]
        optRT = trialsDf[(trialsDf['Block'] == j) & (trialsDf['Trial'] == i) & (trialsDf['Stim Type'] == 'opt')]
        total0sList.append(x0forResponse['Total 0s'].values)
        totalRTOpt.append(optRT['CEDRUS Reaction Time (ms)'].values)
responsesDf['Total 0s'] = total0sList
responsesDf['Opt RT'] = totalRTOpt

responsesDf['Expected 0s'] = responsesDf['Probability of 0'] * 40
responsesDf['Average 0s'] = responsesDf['Total 0s'].map(lambda x: x.mean())
responsesDf
```

```python
responsesDf['totalStimuliDisplay'].sum()
```

```python
probabilityBiasof0 = np.unique(responsesDf['Probability of 0'])
probabilityBiasof0
```

Since theres a 1 second ISI after each response, this is why the total time is approx reaction time + 1 second.

```python
percentCorrect = []
avgSamplesPerTrial = []
avgReactionTimes = []
numberInstances = []
for i in probabilityBiasof0:
    print('Probability:', i)
    df = responsesDf[responsesDf['Probability of 0'] == i]
    print('Number of Instances:', len(df))
    numberInstances.append(len(df))
    print('Percentage Correct', df['Correct'].mean())
    avgSamplesPerTrial.append(df['totalStimuliDisplay'].values.mean())
    print('Avg amount of trials sampled:', df['totalStimuliDisplay'].values.mean())
    
    avgReactionTimes.append(df['CEDRUS Reaction Time (ms)'].values.mean())
    print('Avg amount of time per trial:', df['CEDRUS Reaction Time (ms)'].values.mean())
    display(df)
    percentCorrect.append(df['Correct'].mean())
```

```python
getCounts = lambda probability: [print((responsesDf['Probability of 0'] == p).sum()) for p in probability]

getCounts([0.425, 0.45, 0.55, 0.575])
percentCorrect
```

```python
plt.figure(figsize= (15, 10))
plt.subplot(2, 2, 1)
plt.plot(probabilityBiasof0, percentCorrect, '--*')
plt.xlabel('Probability of Success (45 degree T)')
plt.ylabel('Percentage Correct')

plt.subplot(2, 2, 2)
plt.plot(probabilityBiasof0, avgSamplesPerTrial, '--*')
plt.xlabel('Probability of Success (45 degree T)')
plt.ylabel('Average # of samples per trial')

plt.subplot(2, 1, 2)
plt.hist(responsesDf['Probability of 0'])
plt.xlabel('Probability of Right Tilt T')
plt.ylabel('# of trial occurance')
plt.suptitle('300 Trials')

plt.show()
```
